```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website='https://www.mytek.tn/informatique/ordinateurs-portables/pc-portable.html?product_list_limit=all'
      
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
results= soup.find_all('tr',{'class':'item product product-item product-item-info'})
```

```python
len(results)
```

```python
urls = []
for re in results:
    urls.append(re.find('a',{'class':'product photo product-item-photo'}).get('href'))
urls
    
```

```python

processor=[]
marque=[]
memoire=[]
ecran=[]
disque_dur=[]
carte_graphique=[]
Resolution_Ecran=[]
taille_Ecran=[]
for url in urls:

    resp = requests.get(url)
    soup=BeautifulSoup(resp.content, 'html.parser')
    
    memoire.append(soup.find('td',{'class':'col data','data-th':'Mémoire'}).get_text())
    disque_dur.append(soup.find('td',{'class':'col data','data-th':'Disque Dur'}).get_text())
    carte_graphique.append(soup.find('td',{'class':'col data','data-th':'Carte Graphique'}).get_text())
    r=soup.find('td',{'class':'col data','data-th':'Marque'})
    if r is not None:
        marque.append(r.get_text())
    processor.append(soup.find('td',{'class':'col data','data-th':'Processeur'}).get_text())
    taille_Ecran.append(soup.find('td',{'class':'col data','data-th':'Taille de l écran'}).get_text())
    l=soup.find('td',{'class':'col data','data-th':'Résolution Ecran'})
    if l is not None:
        Resolution_Ecran.append(l.get_text().replace('[','').replace(']',''))


    

    
   
```

```python
price=[]
for re in results:
    price.append(re.find('span',{'class':'price'}).get_text())
    
```

```python
review = {'Marque':marque,'Memoire':memoire,'carte graphique':carte_graphique,'Disque Dur':disque_dur,'Resolution Ecran':Resolution_Ecran,'Taille Ecran':taille_Ecran,'Price':price}
df = pd.DataFrame.from_dict(review, orient='index')
df = df.transpose()
#review=pd.DataFrame({'Marque':marque,'Price':price,'Memoire':memoire,'carte graphique':carte_graphique,'Ecran':ecran,'Disque Dur':disque_dur})
```

```python
df
```

```python
dataset=df.to_csv("Mytekk.csv", encoding="utf-16")
```
