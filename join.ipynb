```python
df_keys['Disque Dur'] = df_keys['Disque Dur'].astype(str).replace('\.0', '', regex=True)
df_keys["Disque Dur"] = df_keys["Disque Dur"].str.replace('Go', '')
df_keys["Disque Dur"] = df_keys["Disque Dur"].str.replace('To', '000')
new = df_keys["Disque Dur"].str.split("+", n = 1, expand = True)

df_keys["first"]= new[0]
df_keys["first"].fillna("0", inplace = True)
df_keys["first"]=df_keys["first"].str.strip()
df_keys["second"]= new[1]
df_keys["Layer1HDD"] = df_keys["first"].apply(lambda x: 1 if "HDD" in x else 0)
df_keys["Layer1SSD"] = df_keys["first"].apply(lambda x: 1 if "SSD" in x else 0)

df_keys["Layer1eMMC"] = df_keys["first"].apply(lambda x: 1 if "eMMC" in x else 0)
df_keys['first'] = df_keys['first'].str.replace(r'\D', '')

df_keys["second"].fillna("0", inplace = True)
df_keys["Layer2HDD"] = df_keys["second"].apply(lambda x: 1 if "HDD" in x else 0)
df_keys["Layer2SSD"] = df_keys["second"].apply(lambda x: 1 if "SSD" in x else 0)
df_keys["Layer2eMMC"] = df_keys["second"].apply(lambda x: 1 if "eMMC" in x else 0)

df_keys['second'] = df_keys['second'].str.replace(r'\D', '')

df_keys["first"] = df_keys["first"].astype(int)
df_keys["second"] = df_keys["second"].astype(int)
df_keys["HDD"]=(df_keys["first"]*df_keys["Layer1HDD"]+df_keys["second"]*df_keys["Layer2HDD"])
df_keys["SSD"]=(df_keys["first"]*df_keys["Layer1SSD"]+df_keys["second"]*df_keys["Layer2SSD"])
df_keys["eMMC"]=(df_keys["first"]*df_keys["Layer1eMMC"]+df_keys["second"]*df_keys["Layer2eMMC"])
df_keys.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 
       'Layer1eMMC', 'Layer2HDD', 'Layer2SSD',
       'Layer2eMMC'],inplace=True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline#output of each step is used as output to the next step
#Pipeline: sequence of transformation(preprocessing,Features Engineering)followed by estimators(Algorithms,models)assembled and excuted as a single entity
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
#Linear regression
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,2,6,7])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
#XGBOOST
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,2,6,7])
],remainder='passthrough')

step2 = XGBRegressor(n_estimators=45,max_depth=5,learning_rate=0.5)#n_estimators: nombre d'arbres que vous voulez construire.
                                                                   #learning_rate : la taille du pas utilisé pour éviter le overfitting. La plage est [0,1]
pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
#Voting Regressor
from sklearn.ensemble import VotingRegressor,StackingRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,2,6,7])
],remainder='passthrough')


rf = RandomForestRegressor(n_estimators=350,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)
gbdt = GradientBoostingRegressor(n_estimators=100,max_features=0.5)
xgb = XGBRegressor(n_estimators=25,learning_rate=0.3,max_depth=5)
et = ExtraTreesRegressor(n_estimators=100,random_state=3,max_samples=0.5,max_features=0.75,max_depth=10)

step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('xgb',xgb), ('et',et)],weights=[5,1,1,1])

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
def save_model_to_db(model,client,db,dbconnection,model_name):
    pickle_model=pickle.dumps(model)
    myclient=pymongo.MongoClient(client)
    mydb=myclient[db]
    mycon=mydb[dbconnection]
    info=mycon.insert_one({model_name:pickle_model,'name':model_name,'created_time':time.time()})
    print(info.inserted_id,'saved with this id succefully!')
    details={
        'inserted_id':info.inserted_id,
        'model_name':model_name,
        'created_time':time.time()
    }
    return details
```

```python
def load_saved_model_from_db(model_name,client,db,dbconnection):
    json_data={}
    myclient=pymongo.MongoClient(client)
    mydb=myclient[db]
    mycon=mydb[dbconnection]
    data=mycon.find({'name':model_name})
    for i in data:
        json_data=i
    pickled_model=json_data[model_name]
    return pickle.loads(pickled_model)
   
```
