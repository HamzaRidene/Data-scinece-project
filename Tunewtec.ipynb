```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website="https://tunewtec.com/c/informatique/pc-et-ordinateur/ordinateur-portable/pc-portable/?orderby=price&per_page=48"
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
results= soup.find_all('div',{'class':"product-wrapper"})
```

```python
len(results)
```

```python
urls = []
for re in results:
    urls.append(re.find({'a'}).get('href'))
len(urls)
    
```

```python
processor=[]
marque=[]
memoire=[]
os=[]
disque_dur=[]
carte_graphique=[]
taille_Ecran=[]
Resolution_Ecran=[]
for url in urls:

    resp = requests.get(url)
    soup=BeautifulSoup(resp.content, 'html.parser')
    processor.append(soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_processeur'}).text.replace('\n','').replace('Processeur',''))
    memoire.append(soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_memoire-ram'}).text.replace('\n','').replace('Mémoire Ram',''))
    disque_dur.append(soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_disque-dur'}).get_text().replace('\n','').replace('Disque Dur',''))
    carte_graphique.append(soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_carte-graphique'}).get_text().replace('\n','').replace('Carte graphique',''))
    marque.append(soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_marque'}).get_text().replace('\n','').replace('Marque',''))
    taille_Ecran.append(soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_taille-de-lecran'}).get_text().replace('\n','').replace('Taille de l’écran','').replace('"',''))
    l=soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_resolution-de-lecran'})
    if l is not None:
        Resolution_Ecran.append(l.get_text().replace('\nRésolution Ecran\n','').replace('\n\n',''))
    os.append(soup.find('tr',{'class':'woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_systeme-dexploitation'}).get_text().replace('\n',''))    
        
    
    
```

```python
price=[]
for re in results:
    price.append(re.find('span',{'class':'price'}).get_text())
    
```

```python
review = {'Marque':marque,'Cpu':processor,'System Exploitation':os,'Memoire':memoire,'carte graphique':carte_graphique,'Disque Dur':disque_dur,'Resolution Ecran':Resolution_Ecran,'Taille Ecran':taille_Ecran,'Price':price}
df = pd.DataFrame.from_dict(review, orient='index')
df = df.transpose()
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df=df.dropna()
```
